PCA

```python
X=dff.copy()
Y=dff['class']
X=X.drop('class',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_train)
```

```python

X_pca = pca.transform(X_train)
pca.components_
pca.explained_variance_
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,7))

plt.plot(np.cumsum(pca.explained_variance_ratio_))

plt.xlabel('Number of components-1')
plt.ylabel('Total explained variance')

plt.xlim(0,9)
plt.xticks(np.arange(0, 9, 1))

plt.axvline(6, c='b')
plt.axhline(0.965, c='r')

plt.show()
```

```python
pca = PCA(n_components=7)
new_data=pca.fit_transform(X_train)
new_data.shape
new_data=pd.DataFrame(new_data,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
#df_scaled=df_scaled.reset_index()

test_data=pca.transform(X_test)
test_data=pd.DataFrame(test_data,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
```

```python
print(new_data.shape)
print(test_data.shape)
```

SVM

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(new_data,y_train)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

prediction = svc.predict(test_data)
## Evaluation
# Accuracy
print("Accuracy:", metrics.accuracy_score(prediction,y_test))
print("Classification Report:\n", metrics.classification_report(prediction,y_test))
print("Confusion Matrix:\n", metrics.confusion_matrix(prediction,y_test))
```
